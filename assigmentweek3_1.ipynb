#### Importing Libraries

```python
import numpy as np #Numpy Library
import pandas as pd #Panda Library
```

```python
from bs4 import BeautifulSoup #BeautifulSoup to read the html
```

```python
import requests #to request the page
```

#### use of request to read the the html page https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M

```python
source=requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text

#print(source)
```

#### use of Beautifulsoup to read the html and obtain the table

```python
soup = BeautifulSoup(source,'lxml')
```

```python
tabla = soup.find('table', attrs={'class':'wikitable sortable'})
print(tabla.prettify())  #prettify to display in a more friendly way
```

#### processing the table to obtain a Pandas dataframe
- use of find_all to obtain the rows and columns of the table
- list is the pandas dataframe with the table info

```python
filas = tabla.find_all('tr')
l=[]
for tr in filas:
    td = tr.find_all('td')
    fila = [tr.text for tr in td]
    l.append(fila)
lista = pd.DataFrame(l, columns=['PostCode','Borough','Neigborhood'])
lista.head()
```

## Cleaning the data
- replace funtion to remove the "\n" of the Neigborhoods

```python
lista = lista.replace('\n','', regex=True)
```

```python
lista.head(10)
```

- removing the first index to has None values

```python
lista.drop(lista.index[0],inplace=True)
```

```python
lista = lista[lista.Borough != 'Not assigned']
lista.head(20)
```

-  aasigning the value of the Borough if the Neigborhood is Not assigned

```python
lista['Neigborhood']= np.where(lista['Neigborhood']=='Not assigned',lista['Borough'],lista['Neigborhood'])
```

```python
print(lista.loc[lista['Neigborhood'] == 'Queen\'s Park']) #checking if the Neigborhood para Queen's Park changed
```

- sorting by the PostCode

```python
lista.sort_values('PostCode',inplace=True)
lista.head(20)
```

- Defining a funtion to concatenate the neigborhoods

```python
concatenar = {
    'Neigborhood': lambda x: ', '.join(x)
}
```

- Grouping the dataframe by postcode and Borough using the concatenar funtion

```python
lista = lista.groupby(['PostCode','Borough'], as_index=False, sort=False).agg(concatenar)
lista.head(20)
```
